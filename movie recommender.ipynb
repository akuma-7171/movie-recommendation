```python
import pandas as pd
import numpy as np
import ast
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')     # reads csv files
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title') #merge both credits and movies files together
```

```python
# pre-processing part start !!!
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']] # choose essential parts/attributes
```

```python
movies.head()
```

```python
movies.isnull().sum()  # check weather dataset contains null value or not
```

```python
movies.dropna(inplace=True) #if it is contains null value then call fuction "dropna" and replace with the original file
```

```python
movies.duplicated().sum() # also check for thr duplicates value
```

```python
movies.iloc[0].genres 
```

```python
def convert(obj):  # write a fuction to convert dict into a simple list
    
    l = []
    
    for i in ast.literal_eval(obj): # it will covert dict into string
        
        l.append(i['name']) # append those string into a list
    
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert) # apply above fuction and replace with original attribute
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert) # apply above fuction and replace with original attribute
```

```python
movies.head()
```

```python
def convert_name(obj): #write a function to fetch first top tree cast name from cast attribute from the dict.
    
    l = []
    
    counter = 0
    
    for i in ast.literal_eval(obj):
        
        if counter != 3:
            
             l.append(i['name']) # append those top three names into the list
                
        else:
            
            break
        
    
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_name) # apply above fuction and replace with original attribute
```

```python
movies.head()
```

```python
def fetch_director(obj): #write a function to fetch director name from crew attribute from the dict.
    
    l = []
    
   
    
    for i in ast.literal_eval(obj):
        
        if i['job'] == 'Director':
            
            l.append(i['name'])
            break
            
               
                
                
        
        
    
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director) # apply above fuction and replace with original attribute
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) # bridge all gapes in overview attribute
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])  # bridge all gapes from all the attributes
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] # merge all the attributes into newly made attribute "tags"
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']] # make new data-frame from the movies dataset
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x)) #make a whole new string for tags attributes
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower()) # convert all the words into lower case
```

```python
from sklearn.feature_extraction.text import CountVectorizer # start text-vectorizer

cv = CountVectorizer(max_features = 5000, stop_words="english") # maximum words 5000 and remove all the stops word
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() #convert tags into array to find the frequency of the words
```

```python
vectors
```

```python
import nltk

from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()  
```

```python

def stem(text): # fuction to find most similar words from dataset
    
    l = []
    
    for i in text.split():
        
        l.append(ps.stem(i))
        
    return " ".join(l)    
```

```python
new_df['tags'] = new_df['tags'].apply(stem) # replace with original attribute after apply the fuction
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarities = cosine_similarity(vectors) # calculate vector distance by using cosine_similarities
```

```python
def recommender(movie):
    
        movie_index = new_df[new_df['title'] == movie].index[0] # fetch the movie index

        distance = similarities[movie_index] # according to movie index it all distance from the other movie

        movie_list = sorted(list(enumerate(distance)), reverse=True, key = lambda x:x[1])[1:6] # grab the input movie and sorted according to most appropriate movie and fetch first five movie


        print("recommend movies for you :- ")
        print(" ")
        print(" ")
        print(" ")


        for i in movie_list:
            print(new_df.iloc[i[0]].title)
    
    
    
   
       
        
        
   
     
     
       
      
```

```python
recommender('John Carter')
```
